# **ÖZEL MODEL KULLANIMI -- CUSTOM MODEL KULLANIMI**

Model içeri aktarımı, tokenize etmek, tahmin işlemi

```python
from transformers import AutoModelForSequenceClassification

# Initializing a model for sequence classification
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
from transformers import AutoTokenizer

# Initializing a tokenizer for the model
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
classifier("C'est un film génial. J'aime ça.")
```

Tokenize işleminin çalışması

```python
encoding = tokenizer("This film is nice. I liked it.")
print(encoding)
```

```python
batch = tokenizer(
    ["I like NLP. This is an awosome area", "We hope you don't hate it."],
    max_length=512,
    truncation=True,
    padding=True,
    return_tensors="pt")
```

```python
batch
```

Modele katman eklemek

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
outputs = model(**batch)
```

```python
from torch import nn

# Applying softmax to model logits to get predictions
predictions = nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

Modeli kaydetmek

```python
save_directory = "./save_pretrained"
tokenizer.save_pretrained(save_directory)
model.save_pretrained(save_directory)
```
